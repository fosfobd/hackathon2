```python
import pandas as pd
import numpy as np
import tensorflow as tf

raw = pd.read_csv("./dataservices_statsocial_raw2.csv", sep=',', header=None)
raw.columns = ['userid', 'affinity']
raw
```

```python
tax = pd.read_csv("./DataServices_StatSocial_Taxonomy.csv", sep=',')
tax.columns = ['affinity', 'lvl1', 'lvl2']
tax
```

```python
merged = pd.merge(raw, tax, on='affinity')
merged
```

```python
merged.sort_values(by=['userid'])
```

```python
uniques = merged.lvl1.unique()
print(uniques)
```

```python
categories = {}
for category in uniques:
    categories[category] = set()
```

```python
for parent, child in zip(merged.lvl1, merged.lvl2): 
    categories[parent].add(child)
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
    
for cat in categories:
    print(color.BOLD + cat + color.END)
    for sub in categories[cat]:
        print('\t', sub)
```
